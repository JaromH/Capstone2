### Data Joining

```python
#Join the dataframes into one large dataframe
merged_features_sales = pd.merge(Features, Sales,  how='left', left_on=['Store','Date'], right_on = ['Store','Date'])
```

```python
merged_features_sales.head()
```

```python
all_merged = pd.merge(merged_features_sales, Stores, how='left', left_on=['Store'], right_on=['Store'])
```

```python
all_merged.head()
```

```python
#Drop the double holiday column
all_merged.drop('IsHoliday_y', axis=1, inplace=True)
```

```python
all_merged.head()
```

From this initial data wrangling, there are no duplicate values and only the features dataframe has missing data. I now want to create two tables, one where the missing markdown data is removed and the other with the missing data replaced by 0.

```python
#Drop markdown columns
all_no_markdown = all_merged.drop(['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'], axis=1)
```

```python
all_no_markdown.head()
```

```python
#Make all NaN markdown values equal 0
all_0_markdown = all_merged
```

```python
all_0_markdown['MarkDown1'] = all_0_markdown['MarkDown1'].fillna(0)
all_0_markdown['MarkDown2'] = all_0_markdown['MarkDown2'].fillna(0)
all_0_markdown['MarkDown3'] = all_0_markdown['MarkDown3'].fillna(0)
all_0_markdown['MarkDown4'] = all_0_markdown['MarkDown4'].fillna(0)
all_0_markdown['MarkDown5'] = all_0_markdown['MarkDown5'].fillna(0)
```
